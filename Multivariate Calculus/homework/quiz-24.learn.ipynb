```python
import numpy as np

# Here the function is defined:
def linfit(xdat, ydat):
    # Here xbar and ybar are the averages
    xbar = np.sum(xdat)/len(xdat)
    ybar = np.sum(ydat)/len(ydat)
    
    # Insert calculation of m an dc here:
    covxy = np.sum((xdat - xbar)*ydat)
    varx = np.sum((xdat - xbar)**2)
    
    m = covxy / varx
    c = ybar - m*xbar
    
    return m, c
```

```python
data = np.array([
    [0.4, 0.1],
    [0.5, 0.25],
    [0.6, 0.55],
    [0.7, 0.75],
    [0.8, 0.85],
])

x = data[:,0]
y = data[:,1]
```

```python
m, c = linfit(x, y)
```

```python
print(m)
print(c)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def line(m=0, c=0, data=data, regression=None):
    x = data[:, 0]
    y = data[:, 1]
    if regression is not None:
        m = regression.slope
        c = regression.intercept
    y_pred = m*x + c
#     print(x)
#     print(y)
#     print(y_pred)

    plt.plot(x, y_pred, '-r', label="y=mx+c")
    plt.scatter(x, y, marker='o', color='blue')
    plt.title('Graph of y=mx+c')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='upper left')
    plt.grid()
    plt.show()
```

```python
line(m, c)
```

```python
from scipy import stats

# Use the stats.linregress() method to evaluate regression
regression = stats.linregress(x, y)

line(regression=regression)
```
